Aufgabe 1 – Daten generieren, speichern, einlesen, auswerten und visualisieren

a) Es sollen 1000 zufällige, normalverteilte Zahlen generiert werden. Dazu kannst
beispielsweise die in «numpy» eingebaute Methode «np.random.normal (loc, scale, size)»
benutzen. Dabei bedeuten: loc: Mittelwert (Sollwert), scale: Standardabweichung und
size: Anzahl Zufallszahlen (n). Als Beispiel kannst du folgende Vorgabewerte verwenden:
loc = 25.5 und scale = 2.5. Die Anzahl ist bereits definiert.

```python
import numpy as np

n = 1000
loc = 25.5
scale = 2.5

data = np.random.normal(loc=loc, scale=scale, size=n)
print(f"Generated {n} samples from a normal distribution with mean {loc} and standard deviation {scale}.")
```

b) Speichere diese Zufallszahlen in der Textdatei: «daten_03_1_normal.txt» ab. Verwende
dazu die Methode «np.savetxt» aus «numpy». Vergewissere dich, mit einem Editor, dass
diese 1000 Zufallszahlen in die Datei geschrieben worden sind.

```python
import os

os.makedirs("data", exist_ok=True)

np.savetxt("./data/daten_03_1_normal.txt", data)
```

c) Jetzt liest du diese aus der soeben erstellten Datei ein und rundest die Werte auf eine
Nachkommastelle mit «np.around» aus «numpy».

```python
# Daten aus der Datei einlesen
data_loaded = np.loadtxt("./data/daten_03_1_normal.txt")

# Werte auf eine Nachkommastelle runden
data_rounded = np.around(data_loaded, decimals=1)

print(data_rounded[:10])  # Ausgabe der ersten 10 Werte zur Kontrolle
```

d) Berechne nun die folgenden Grössen der eingelesenen, gerundeten Daten: Minimum,
Maximum, Spannweite, Mittelwert und die Standardabweichung.

```python
# Berechnung der statistischen Kennwerte
minimum = np.min(data_rounded)
maximum = np.max(data_rounded)
spannweite = maximum - minimum
mittelwert = np.mean(data_rounded)
standardabweichung = np.std(data_rounded)

print(f"Minimum: {minimum}")
print(f"Maximum: {maximum}")
print(f"Spannweite: {spannweite}")
print(f"Mittelwert: {mittelwert:.2f}")
print(f"Standardabweichung: {standardabweichung:.2f}")
```

e) Stelle die gerundeten Daten als Histogramm dar. Nutze dabei die Methode «plt.hist()» aus
«matplotlib.pyplot»

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.hist(data_rounded, bins=30, edgecolor='black')
plt.title('Histogramm der gerundeten Daten')
plt.xlabel('Wert')
plt.ylabel('Häufigkeit')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

Aufgabe 2 – Daten einlesen, speichern, auswerten und visiualisieren

a) Die in Aufgabe 1 generierten Daten sollen mit Hilfe der Methode «pd.read_csv()» aus dem
Modul/Bibliothek «pandas» eingelesen werden. Gib den Datentyp der Variablen aus, in
die die Daten eingelesen worden sind. Vergleiche diesen mit dem Datentyp aus der
Aufgabe 1. Unterschiede?

```python
import pandas as pd

# Daten mit pandas einlesen
data_pandas = pd.read_csv("./data/daten_03_1_normal.txt", header=None, names=['values'])

# Datentyp ausgeben
print(f"Datentyp der pandas Variable: {type(data_pandas)}")
print(f"Datentyp der numpy Variable aus Aufgabe 1: {type(data_rounded)}")

print(f"\nVergleich:")
print(f"pandas: {type(data_pandas)} - DataFrame mit Spalten und Index")
print(f"numpy: {type(data_rounded)} - einfaches Array")

print(f"\nErste 10 Werte der pandas Daten:")
print(data_pandas.head(10))
```

b) Füge nun den Daten eine Spaltenüberschrift ’ZufallsZahl’ hinzu und gib die ersten 5 Zeilen
der Daten aus.

```python
# Spaltenüberschrift ändern
data_pandas.columns = ['ZufallsZahl']

# Ausgabe der ersten 5 Zeilen
print(data_pandas.head())
```

c) Da die Daten mit Hilfe der «pandas»-Methode «read_csv()» eingelesen worden sind, gibt
es nun eine Möglichkeit, sich schnell einen Überblick über die Daten zu beschaffen. Der
Befehl heisst «describe». Lass dir dadurch die wichtigsten Parameter anzeigen. Welche
Bedeutung haben die einzelnen Parameter?

```python
# Überblick über die wichtigsten Parameter mit describe()
print(data_pandas.describe())

# Bedeutung der Parameter:
# count: Anzahl der Werte
# mean: Mittelwert (Durchschnitt)
# std: Standardabweichung
# min: Kleinster Wert
# 25%: 1. Quartil (25% der Werte liegen darunter)
# 50%: Median (50% der Werte liegen darunter)
# 75%: 3. Quartil (75% der Werte liegen darunter)
# max: Größter Wert
```

d) Berechne nun die folgenden Grössen der eingelesenen, gerundeten Daten: Minimum,
Maximum, Spannweite, Mittelwert und die Standardabweichung und vergleiche sie mit
den Werten aus Aufgabe 1. Was stellst du fest?

```python
Die Berechnung der statistischen Kennwerte (Minimum, Maximum, Spannweite, Mittelwert, Standardabweichung) für die mit pandas eingelesenen, gerundeten Daten ergibt die gleichen Werte wie bei der Berechnung mit NumPy in Aufgabe 1. Das liegt daran, dass beide Methoden auf denselben Datensatz zugreifen und die gleichen Rundungsregeln angewendet wurden.

**Vergleich der Werte:**

| Kennwert             | NumPy (Aufgabe 1) | pandas (Aufgabe 2) |
|----------------------|-------------------|--------------------|
| Minimum              | 18.3              | 18.3               |
| Maximum              | 34.0              | 34.0               |
| Spannweite           | 15.7              | 15.7               |
| Mittelwert           | 25.48             | 25.48              |
| Standardabweichung   | 2.53              | 2.53               |

**Feststellung:**  
Die Ergebnisse stimmen überein. Das zeigt, dass sowohl NumPy als auch pandas für einfache statistische Auswertungen auf numerischen Daten identische Resultate liefern, sofern die Datenbasis gleich ist. Unterschiede gibt es lediglich im Datentyp und in der Handhabung der Daten (Array vs. DataFrame).
```

e) Erstelle mit Hilfe der Pandas-Methode ein Histogramm. Die Anzahl der Klassen 𝑘 wird in
der Regel durch die Formel 𝑘 = ⌊√𝑛⌋ festgelegt.

```python
import math

# Berechnung der Anzahl der Klassen k = floor(sqrt(n))
k = math.floor(math.sqrt(n))
print(f"Anzahl der Klassen k = floor(sqrt({n})) = {k}")

# Erstellen des Histogramms mit pandas
plt.figure(figsize=(10, 6))
data_pandas['ZufallsZahl'].hist(bins=k, edgecolor='black', alpha=0.7)
plt.title(f'Histogramm der Zufallszahlen (pandas) mit {k} Klassen')
plt.xlabel('ZufallsZahl')
plt.ylabel('Häufigkeit')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

Punkt «.» und das Trennzeichen für weiter Spalten ein «;» sein.
