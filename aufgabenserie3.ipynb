Aufgabe 3 – Daten generieren, visualisieren und vergleichen

a) Erstelle durch die Methode «np.random.normal(0,1,1000)» 1000 Zufallszahlen mit dem
Mittelwert 0 und der Standardabweichung 1. Damit wir jedes Mal die gleichen Werte
erhalten, kannst du vorher, einen Initialwert für den Zufallszahlen-Generator festlegen
«np.random.seed(0)»

```python
# Initialwert für den Zufallszahlen-Generator festlegen
np.random.seed(0)

# 1000 Zufallszahlen mit Mittelwert 0 und Standardabweichung 1 generieren
data_standard = np.random.normal(0, 1, 1000)

print(f"Generated {len(data_standard)} samples from a standard normal distribution (mean=0, std=1).")
print(f"First 10 values: {data_standard[:10]}")
```

b) Die Zufallszahlen, welche wir generiert haben, sollten normalverteilt sein. Was heisst das?
Es gibt mehrere Möglichkeiten, von Daten festzustellen, ob sie normalverteilt sind oder
 nicht. Eine visuelle Möglichkeit ist der Q-Q-Plot. Erstelle mit der Methode «sm.qqplot()»
ein solches Diagramm. Falls die Messpunkte angenähert eine Gerade darstellen, dann sind
die Messwerte normalverteilt.
Hinweis: Es kann sein, dass du das Modul «statsmodels» zuerst installieren musst. Das
heisst, du kannst in VS Code das Terminal-Fenster anzeigen und dort den Befehl «pip install
statsmodels» eingeben und ausführen. Danach sollte ein Import im JupyterNotebook
möglich sein.

```python
import statsmodels.api as sm

# Q-Q-Plot erstellen
fig, ax = plt.subplots(figsize=(8, 6))
sm.qqplot(data_standard, line='s', ax=ax)
plt.title('Q-Q-Plot für Normalverteilung')
plt.xlabel('Theoretische Quantile')
plt.ylabel('Stichproben-Quantile')
plt.grid(True, alpha=0.3)
plt.show()

print("Q-Q-Plot Interpretation:")
print("- Falls die Punkte annähernd auf einer Geraden liegen, sind die Daten normalverteilt")
print("- Abweichungen von der Geraden zeigen Abweichungen von der Normalverteilung an")
print("- Die rote Linie zeigt die theoretische Normalverteilung")
```

c) Ergänze in deinem Q-Q-Plot noch eine Gerade mit der Steigung 1 (Steigungswinkel ist 45°).
Dies kannst du erreichen, indem du den Befehl «sm.qqplot()» durch einen Parameter
erweiterst.

```python
# Q-Q-Plot mit Gerade der Steigung 1 (45° Winkel)
fig, ax = plt.subplots(figsize=(8, 6))
sm.qqplot(data_standard, line='45', ax=ax)
plt.title('Q-Q-Plot für Normalverteilung mit 45°-Linie')
plt.xlabel('Theoretische Quantile')
plt.ylabel('Stichproben-Quantile')
plt.grid(True, alpha=0.3)
plt.show()

print("Q-Q-Plot mit 45°-Linie (Steigung 1):")
print("- Die Linie zeigt die perfekte Normalverteilung mit Steigung 1")
print("- Parameter 'line='45'' fügt eine Gerade mit 45° Steigungswinkel hinzu")
print("- Je näher die Punkte an dieser Linie liegen, desto normalverteilter sind die Daten")
```

Aufgabe 6 – Daten auswerten und vergleichen (Zeichenhäufigkeiten) - (LINK)

In der Datei «daten_03_6.txt» sind deutsche Textabschnitte gespeichert.

a) Lies diese Daten ein und zerlege diese in einzelne Zeichen, so dass 
du eine Auswertung
über die Häufigkeiten (absolute und relative Anzahl) erstellen kannst. Dabei sollst du
keine Unterscheidung zwischen Klein- und Grossbuchstaben berücksichtigen.
Sonderzeichen kannst du ebenfalls ignorieren.

```python
import re

# Deutsche Textdaten einlesen
with open("./daten/daten_03_6.txt", "r", encoding="utf-8") as file:
    text = file.read()

# Text in Kleinbuchstaben umwandeln und nur Buchstaben beibehalten
text_lower = text.lower()
# Nur deutsche Buchstaben (a-z, ä, ö, ü, ß) extrahieren
only_letters = re.sub(r'[^a-zäöüß]', '', text_lower)

# Zeichen in numpy array umwandeln für weitere Verarbeitung
chars_array = np.array(list(only_letters))

# Eindeutige Zeichen und ihre Häufigkeiten ermitteln
unique_chars, counts = np.unique(chars_array, return_counts=True)

# Relative Häufigkeiten berechnen
total_chars = len(chars_array)
relative_freq = counts / total_chars

# Ergebnisse in DataFrame für bessere Darstellung
results_df = pd.DataFrame({
    'Zeichen': unique_chars,
    'Absolute_Häufigkeit': counts,
    'Relative_Häufigkeit': relative_freq,
    'Prozent': relative_freq * 100
})

# Nach absoluter Häufigkeit sortieren (absteigend)
results_df = results_df.sort_values('Absolute_Häufigkeit', ascending=False)

print(f"Gesamtanzahl der analysierten Zeichen: {total_chars}")
print(f"Anzahl unterschiedlicher Zeichen: {len(unique_chars)}")
print("\nZeichenhäufigkeiten (sortiert nach absoluter Häufigkeit):")
print(results_df.to_string(index=False, float_format='%.4f'))
```

b) Schreibe die einzelnen Zeichen und deren Häufigkeiten in eine Excel-Datei mit dem Namen
«daten_03
_
6_out.xlsx». Vergleiche diese Auswertung mit den Angaben auf der Seite,
welche via obigem Link erreichbar ist. Kannst du bestätigen, dass es sich um deutschen
Text handelt?

```python
# Excel-Datei mit den Zeichenhäufigkeiten erstellen
results_df.to_excel("./data/daten_03_6_out.xlsx", index=False, sheet_name='Zeichenhäufigkeiten')

print("Zeichenhäufigkeiten wurden in 'daten_03_6_out.xlsx' gespeichert.")

# Vergleich mit typischen deutschen Buchstabenhäufigkeiten
print("\nVergleich mit typischen deutschen Buchstabenhäufigkeiten:")
print("Die häufigsten Buchstaben im Deutschen sind normalerweise: E, N, I, S, R, A, T, D, H, U")
print(f"In unserem Text sind die häufigsten: {', '.join(results_df.head(10)['Zeichen'].str.upper())}")

# Analyse der charakteristischen deutschen Buchstaben
german_chars = results_df[results_df['Zeichen'].isin(['ä', 'ö', 'ü', 'ß'])]
if not german_chars.empty:
    print(f"\nDeutsche Umlaute und ß gefunden:")
    print(german_chars[['Zeichen', 'Prozent']].to_string(index=False, float_format='%.2f'))
    print("Dies bestätigt, dass es sich um deutschen Text handelt.")

# Prüfung der häufigsten Buchstaben
top_5_chars = results_df.head(5)['Zeichen'].tolist()
if 'e' in top_5_chars and 'n' in top_5_chars:
    print(f"\nDie Buchstaben 'E' und 'N' sind unter den Top 5 ({top_5_chars}), was typisch für deutsche Texte ist.")
```

```python
# Funktion zum Einlesen und Kategorisieren der Würfeldaten
def wuerfel_daten_auswerten(dateipfad):
    # Daten einlesen, Header überspringen
    daten = np.genfromtxt(dateipfad, delimiter=';', skip_header=1)
    # Die erste Spalte ist die Rundennummer, die restlichen Spalten sind die Würfelergebnisse
    augenzahlen = daten[:, 1:]
    # Gesamtanzahl Würfe pro Runde
    gesamt_wuerfe_pro_runde = np.sum(augenzahlen, axis=1)
    # Gesamtanzahl jeder Augenzahl über alle Runden
    gesamt_augen = np.sum(augenzahlen, axis=0)
    # Relative Häufigkeit jeder Augenzahl
    rel_haeufigkeit = gesamt_augen / np.sum(gesamt_augen)
    return {
        "gesamt_wuerfe_pro_runde": gesamt_wuerfe_pro_runde,
        "gesamt_augen": gesamt_augen,
        "rel_haeufigkeit": rel_haeufigkeit
    }

# Beispiel für eine Datei (für alle 5 Dateien analog anwenden)
dateien = [
    "./daten/daten_03_7_1.txt",
    "./daten/daten_03_7_2.txt",
    "./daten/daten_03_7_3.txt",
    "./daten/daten_03_7_4.txt",
    "./daten/daten_03_7_5.txt"
]

auswertungen = []
for datei in dateien:
    auswertung = wuerfel_daten_auswerten(datei)
    auswertungen.append(auswertung)

# Ergebnisse für die erste Datei anzeigen
print("Gesamtanzahl Würfe pro Runde (Datei 1):", auswertungen[0]["gesamt_wuerfe_pro_runde"])
print("Gesamtanzahl jeder Augenzahl (Datei 1):", auswertungen[0]["gesamt_augen"])
print("Relative Häufigkeit jeder Augenzahl (Datei 1):", np.round(auswertungen[0]["rel_haeufigkeit"], 3))
```

```python
# c) Visualisierung der relativen Häufigkeiten jeder Augenzahl für alle 5 Würfel als Säulendiagramme

fig, axes = plt.subplots(1, 5, figsize=(20, 4), sharey=True)
augen_labels = ['EINS', 'ZWEI', 'DREI', 'VIER', 'FUENF', 'SECHS']

for i, auswertung in enumerate(auswertungen):
    axes[i].bar(augen_labels, auswertung["rel_haeufigkeit"], color='skyblue', edgecolor='black')
    axes[i].set_title(f"Würfel {i+1}")
    axes[i].set_xlabel("Augenzahl")
    if i == 0:
        axes[i].set_ylabel("Relative Häufigkeit")
    axes[i].set_ylim(0, 0.4)
    for j, v in enumerate(auswertung["rel_haeufigkeit"]):
        axes[i].text(j, v + 0.01, f"{v:.2f}", ha='center', fontsize=9)

plt.suptitle("Relative Häufigkeiten der Augenzahlen für alle 5 Würfel")
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()

# d) Analyse: Welche Würfel waren gezinkt?
# Ein fairer Würfel hätte für jede Augenzahl eine relative Häufigkeit von etwa 1/6 ≈ 0.167.
# Wir markieren Würfel, bei denen eine Augenzahl deutlich häufiger oder seltener als erwartet auftritt.

gezinkt = []
for idx, auswertung in enumerate(auswertungen):
    if np.any(auswertung["rel_haeufigkeit"] > 0.2) or np.any(auswertung["rel_haeufigkeit"] < 0.05):
        gezinkt.append(idx + 1)

if gezinkt:
    print(f"Die folgenden Würfel zeigen Auffälligkeiten und könnten gezinkt sein: {gezinkt}")
else:
    print("Alle Würfel zeigen eine annähernd gleichmäßige Verteilung und sind vermutlich fair.")
```
